```python
def fuck_P(P):
    fuck = []
    for p in P:
        fuck.append(A[p])
    return fuck
```

```python
def fuck_top5(top5):
    fuck_5 = []
    for i in top5:
        fuck_temp = []
        for j in i:
            fuck_temp.append(A[j])
        fuck_5.append(fuck_temp)
    return fuck_5
```
